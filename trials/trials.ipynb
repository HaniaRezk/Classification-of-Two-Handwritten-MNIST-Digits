## Loading the test data

```python
loaded_y_test_array = np.load("test_y_data.npy")
loaded_x_test_array = np.load("test_x_data.npy")
print("Testing data loaded")

```

# Loading the best neural network and its weights.

```python
#Load the best model
best_model_loaded = load_model("NN_files/best_model.h5")
best_model_loaded.load_weights('NN_files/best_model.weights.h5')
print("Best neural network loaded!")
print(best_model_loaded.summary())
```

```python
weights = best_model_loaded.get_weights()
print("The weights of our best model:")
print("\n")
for i, weight in enumerate(weights):
    print(f"Weight {i}: {weight}")
```

```python
#Evaluating the best model on test data
best_score_loaded= best_model_loaded.evaluate(loaded_x_test_array, loaded_y_test_array, verbose=0)
print("Test accuracy for the best model loaded from the file is:",best_score_loaded[1])
```

```python
#Load the history
with open('NN_files/best_training_history.json', 'r') as f:
    loaded_history_best = json.load(f)
print("Training history of best neural network loaded!")
```

Plotting the training and validation accuracies of the best model across epochs

```python
#Plot the Validation, Training and Testing accuracy
plt.plot(loaded_history_best ['accuracy'][5:], label='Training Accuracy')
plt.plot(loaded_history_best ['val_accuracy'][5:], label='Validation Accuracy')
plt.axhline(y=best_score_loaded[1], color='r', linestyle='--', label='Test accuracy')
plt.title('Training, Validation and Testing Accuracies of the Best Neural Network')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```

##Loading the base neural network and its weights.

```python
base_model_loaded = load_model("NN_files/base_model.h5")
base_model_loaded.load_weights('NN_files/base_model.weights.h5')
print("Base neural network loaded!")
print(base_model_loaded.summary())
```

```python
#Loading the weights
weights_base=base_model_loaded.get_weights()
print("The weights of our base model:")
print("\n")
for i, weight in enumerate(weights_base):
    print(f"Weight {i}: {weight}")
```

```python
#Load the history
with open('NN_files/base_training_history.json', 'r') as f:
    loaded_history_base = json.load(f)
print("Training history of base neural network loaded!")
```

Plotting the training and validation accuracies of the base model across epochs

```python
#Extract accuracy and validation accuracy from the history object
base_train_accuracy = loaded_history_base['accuracy']
base_val_accuracy =loaded_history_base['val_accuracy']

#Plot training and validation accuracy
plt.plot(range(1, 21), base_train_accuracy, label='Training Accuracy')
plt.plot(range(1, 21), base_val_accuracy, label='Validation Accuracy')
plt.title('Training and Validation Accuracy of the Base Neural Network')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.grid(True)
plt.show()
```
